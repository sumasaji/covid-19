# **Visualization for Customer Behavior Analysis**

```python
sns.histplot(x='Income', data=df, color="darkblue",kde=1)
plt.title("Distribution of Income Variable")
plt.xlabel('Income')
plt.ylabel('Frequency')
plt.show()
```

```python
spending_cols = ['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']
df['total_spending'] = df[spending_cols].sum(axis=1)

```

```python
df[spending_cols].sum().plot(kind='bar', color='purple')
plt.title('Total Spending by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Total Spending')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.countplot(x='Education', data=df, palette='plasma')
plt.title("Customer Distribution by Education")
plt.show()
sns.countplot(x='Marital_Status', data=df, palette='viridis')
plt.title('Customer Distribution by Marital Status')
plt.show()
```

```python
purchase_cols=['NumDealsPurchases','NumWebPurchases','NumCatalogPurchases','NumStorePurchases','NumWebVisitsMonth']
df[purchase_cols].sum().plot(kind='bar',color='teal')
plt.title("Total Purchases by Channel")
plt.xlabel('Channel')
plt.ylabel('Total Purchases')
plt.xticks(rotation=45)
plt.show()
```

```python
campaign_cols=['AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4','AcceptedCmp5',]
df[campaign_cols].sum().plot(kind='bar',color='orange')
plt.title("Response to Marketing Campaigns")
plt.xlabel("Campaign")
plt.ylabel("Number of Responses")
plt.xticks(rotation=45)
plt.show()
```

### ***Customer Segmentation with Spending***

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(x='Income',
                y='total_spending',
                hue='Marital_Status',
                data=df,
                palette='muted',
                style='Marital_Status',
                markers=True,
                s=100)

plt.title("Income vs. Total Spending by Marital Status")
plt.xlabel("Income")
plt.ylabel("Total Spending")
plt.legend(title='Marital Status', loc='upper right')
plt.show()
```

# *** Analyzing Customer Recency***
